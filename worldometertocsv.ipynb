```python
import requests
import csv
from bs4 import BeautifulSoup
url="https://www.worldometers.info/coronavirus/"
page=requests.get(url)
page=page.text

content=BeautifulSoup(page,"html.parser")
el=[]
all_div=content.find_all("tr")
for i in all_div:
    #print(type(i.text))
    #print(i.text)
    x=i.text.split()
    el.append(x)
print(el)
with open("worldometer.csv","w",newline='') as file:
        writer=csv.writer(file)
        writer.writerows(el)
    
    
   
        
   
```
