```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from datetime import datetime
import pandas as pd
driver=webdriver.Chrome('C:/Users/LENOVO/Desktop/chromedriver.exe')
driver.get('https://web.whatsapp.com/')




t_list=['"Papa"','"Natasha Di"','"Shreyansh"','"Mummy"','"Khushi 2"']


online_list=[]
offline_list=[]
try:
    for i in range(0,len(t_list)):
        wait=WebDriverWait(driver,30)
        x_arg='//span[contains(@title,' + t_list[i] + ')]'
        print(x_arg)
        target=wait.until(EC.presence_of_element_located((By.XPATH,x_arg)))
#         print(target.text)
        target.click()


        o_target='"online"'
        o_arg='//span[contains(@title,'+ o_target +')]'
    
        try:
            online_target=wait.until(EC.presence_of_element_located((By.XPATH,o_arg)))
            if (online_target.text=="online"):
                x=str(datetime.now()).split()
                x=x[1].split(".")
                online_list.append(x[0])
                while(online_target.text=="online"):
                    
                    online_target=wait.until(EC.presence_of_element_located((By.XPATH,o_arg)))
                
                
        except Exception as e:
            y=str(datetime.now()).split()
            y=y[1].split(".")
            offline_list.append(y[0])
            online_list.append("none")
            
            
            
            
except Exception as e:
     print("offline")
     print(f"error is {e}")
     


print(online_list)
print(offline_list)
df=pd.DataFrame(list(zip(online_list,offline_list,)),index=t_list,columns=['online','offline'])
df













```
